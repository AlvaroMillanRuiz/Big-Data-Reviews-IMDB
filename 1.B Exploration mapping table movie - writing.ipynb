# Exporation mapping table movie - writing

```python
import pandas as pd
import duckdb
from collections import Counter
import numpy as np
```

```python
# Load mapping table movie to director
mapping_movie_writer = pd.read_json('data/writing.json')

# Replace missing values by legit missing values
mapping_movie_writer = mapping_movie_writer.replace('\\N', np.nan)
mapping_movie_writer = mapping_movie_writer.dropna()

mapping_movie_writer
```

```python
# Count the writers per movie, in descending order
mapping_movie_writer.groupby('movie').count().sort_values('writer', ascending=False)
```

```python
# My suggestion is to only select first select top 5 directors
grouped_mapping_movie_writer = mapping_movie_writer.groupby(by='movie').count().sort_values(by='writer', ascending=False)
Counter(grouped_mapping_movie_writer['writer'])
```

```python
# Select first 8 Concat writers per movie
top_8_writers_per_movie = mapping_movie_writer.groupby('movie').head(8).groupby(['movie'], as_index=False).agg({'writer': ', '.join})
movie_and_writers = pd.concat([top_8_writers_per_movie['movie'], top_8_writers_per_movie['writer'].str.split(', ', expand=True)], axis=1)
movie_and_writers.rename(columns={0: "writer 1", 1: "writer 2", 2: "writer 3", 3: "writer 4", 4: "writer 5", 5: "writer 6", 6: "writer 7", 7: "writer 8"})
```
